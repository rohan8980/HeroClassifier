```python
from bs4 import BeautifulSoup
import requests
from PIL import Image
import os
import base64
```

```python
def get_image_links_from_url(url):
        images_links = []
        
        # Getting html response from the url
        response = requests.get(url).text
        page = BeautifulSoup(response, 'lxml')
            #print(page)
            # formatted_html = page.prettify()
            # print(formatted_html)

        # Getting all the img tags from the page
        images = page.find_all('img')
            #print (len(images))

        # For each tag fetch the src and save it in image_urls dictonary for that superhero
        for image in images:
            # logic for google images
            if 'google.com' in url:
                image_url = image.get('src')

            # logic for wallpapers.com
            if 'wallpapers.com' in url:
                image_url = image.get('data-src')
                #print(image_url)
                if image_url:
                    image_url = 'https://wallpapers.com/'+image_url
            
            # logic for deviantart.com
            if 'deviantart.com' in url:
                image_url = image.get('src')
                #print(image_url)
                if '.deviantart.net' in image_url:  #also has some base64 encoded images
                    continue
            
            #print(image_url)
            if image_url:
                images_links.append(image_url)
                    
        return images_links


def save_image(url, filepath):
    try:
        # base64 encoded images
#         if url.startswith("data:"):
#             #url = 'data:image/png;base64,iVBORw0KGgoAAAANSUhEUgAAAAEAAAABCAQAAAC1HAwCAAAAC0lEQVR42mNkYAAAAAYAAjCB0C8AAAAASUVORK5CYII='

#             # Fetching extension
#             extension = url.split(';')[0].split('image/')[1]
#             filepath = filepath+'.'+extension
#             print(filepath)
#             if os.path.exists(filepath):
#                 print(f'Already Exists {filepath}')
#                 #return

#             # Remove any data URI header if present
#             url = url.split(",")[1]

#             # Decode the base64 data and save as an image file
#             image_bytes = base64.b64decode(url)
#             with open(filepath, 'wb') as f:
#                 f.write(image_bytes)
        
#         # images with links
#         else:
        # Get request to download image
        response = requests.get(url, stream=True)
        response.raise_for_status()

        # Fetching the extension of the image from the response
        content_type = response.headers.get("Content-Type")
        if content_type and content_type.startswith("image/"):
            extension = content_type.split("/")[-1].lower()
        else:
            extension = "jpeg" 

        # Saving image to filepath location
        filepath = filepath+'.'+extension
        if os.path.exists(filepath):
                #print(f'Already Exists {filepath}')
            return
        with open(filepath, 'wb') as f:
            for chunk in response.iter_content(1024):
                f.write(chunk)

        return filepath
        
    except requests.exceptions.RequestException as e:
        msg = '/'.join(filepath.split('\\')[-2:])
        #print(f"Error downloading image: {filepath} - {url[:20]}") #{url[:20]} \n error: {e}")
        print(f"Error: {msg} --> {url}")
        #return [folder, url, e]
        
        
def resize_with_same_aspect_ratio_and_padding(filepath, final_size = (224, 224)):
    padding_colour = (0, 0, 0)
    with Image.open(filepath) as image:
        width, height = image.size

        # Calculate new dimensions while maintaining aspect ratio
            #new_width, new_height = final_size
        sizepx = 224
        new_width = sizepx 
        new_height = sizepx

        if width / height > new_width / new_height:
          # Wider image: scale based on width
          new_height = int(height * (new_width / width))
        else:
          # Taller image: scale based on height
          new_width = int(width * (new_height / height))

        # Resize the image with adjusted dimensions
        resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)

        # Add padding if necessary
        
        #if new_width != final_size[0] or new_height != final_size[1]:
            #left = (final_size[0] - new_width) // 2
            #top = (final_size[1] - new_height) // 2
        if new_width != sizepx or new_height != sizepx:
            left = (sizepx - new_width) // 2
            top = (sizepx - new_height) // 2
            right = left + new_width
            bottom = top + new_height
            padding_image = Image.new("RGB", final_size, padding_colour)  # Set background color for padding
            padding_image.paste(resized_image, (left, top))
            resized_image = padding_image

        # Save the resized image
        resized_image.save(filepath)
        
        
def print_fetched_image_links(image_urls):
    for superhero in image_urls:
        print(superhero, '-', len(image_urls[superhero]))
    for superhero, value in image_urls.items():
        print('\n',superhero, '-', len(value))
        for url in value:
            print('-->', url)                            
```
