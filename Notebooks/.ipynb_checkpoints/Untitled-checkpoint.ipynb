```python
import tensorflow as tf
from PIL import Image

def preprocess_image(img_path, img_size):
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=img_size)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.keras.applications.vgg16.preprocess_input(img_array)
    return img_array
def resize_with_same_aspect_ratio_and_padding(filepath, final_size = (224, 224)):
    padding_colour = (255, 255, 255)
    with Image.open(filepath) as image:
        width, height = image.size

        # Calculate new dimensions while maintaining aspect ratio
            #new_width, new_height = final_size
        sizepx = 224
        new_width = sizepx 
        new_height = sizepx

        if width / height > new_width / new_height:
          # Wider image: scale based on width
          new_height = int(height * (new_width / width))
        else:
          # Taller image: scale based on height
          new_width = int(width * (new_height / height))

        # Resize the image with adjusted dimensions
        resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)

        # Add padding if necessary
        
        #if new_width != final_size[0] or new_height != final_size[1]:
            #left = (final_size[0] - new_width) // 2
            #top = (final_size[1] - new_height) // 2
        if new_width != sizepx or new_height != sizepx:
            left = (sizepx - new_width) // 2
            top = (sizepx - new_height) // 2
            right = left + new_width
            bottom = top + new_height
            padding_image = Image.new("RGB", final_size, padding_colour)  # Set background color for padding
            padding_image.paste(resized_image, (left, top))
            resized_image = padding_image

        # Save the resized image
        resized_image.save(filepath)
```

```python
import os
import tensorflow as tf
from PIL import Image

base_path = r'D:\Git\HeroClassifier\Notebooks\UI Images'

superheroes = {
    'Batman': 'Batman.jpg',
    'Superman': 'superman.webp',
    'Wonder Woman': 'Wonder Woman.jpg',
    'The Flash': 'The Flash.webp',
    'Black Panther': 'Black Panther.png',
    'Spiderman': 'spiderman.jpg',
    'Iron Man': 'Iron Man.webp',
    'Captain America': 'Captain America.jpg',
    'Hulk': 'Hulk.jpg',
    'Black Widow': 'Black Widow.jpg',
}

img_paths = [os.path.join(base_path, name) for name in superheroes.values() ]

  
for img_path in img_paths:
#     image = Image.open(img_path)
#     new_image = image.resize((100, 100))
    resize_with_same_aspect_ratio_and_padding(img_path, (224,224))
```
